```python
from math import sqrt
a,b, c = map(int, input("Enter the coefficients of quadratic equation: ").split())
D = (b**2) - (4*a*c)
if D> 0:
    print("Two distinct real roots exist: ")
    r1 = (-b + sqrt(D)) / (2*a)
    r2 = (-b - sqrt(D)) / (2*a)
    print(f"Root1 = {r1} and Root2 = {r2}")
elif D == 0:
    print("Only one root exist")
    r = -b / (2*a)
    print(f"Root = {r}")
else:
    print("Roots are imaginary.")
    real_part = -b / (2*a)
    imaginary_part = sqrt(abs(D)) / (2*a)
    print(f"Root 1 = {real_part} + {imaginary_part}i")
    print(f"Root 2 = {real_part} + {imaginary_part}i")
```
